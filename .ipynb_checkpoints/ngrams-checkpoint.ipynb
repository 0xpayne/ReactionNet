```python
def smi_tokenizer(smi):
    smi = smi.rstrip()
    pattern =  "(\[[^\]]+]|Br?|Cl?|N|O|S|P|F|I|b|c|n|o|s|p|\(|\)|\.|=|#|-|\+|\\\\|\/|:|~|@|\?|>|\*|\$|\%[0-9]{2}|[0-9])"
    regex = re.compile(pattern)
    tokens = [token for token in regex.findall(smi)]
    sanitycheck = ''.join(tokens)
    assert smi == str(sanitycheck)
    return ' '.join(tokens)
```
